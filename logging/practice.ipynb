7. Log an error message when a division by zero occurs. 


```python
def logg():
    logging.basicConfig(
        level=logging.DEBUG,
        handlers=[
            logging.FileHandler('app7.log'),
        ],
        force=True
    )
    def divide(a,b):
        try:
            result = a/b
            logging.debug(f'Division = {a} / {b} = {result}')
        
        except ZeroDivisionError:
            logging.error('Can not divide by zero')
    divide(10,0)
    divide(4,5)
logg()

```

1. Configure RotatingFileHandler to:
- Create app.log
- Rotate the log file when it exceeds 1 MB
- Keep 5 backup files

```python
from logging.handlers import RotatingFileHandler
logging.basicConfig(
    level = logging.DEBUG,
    format='%(asctime)s - %(levelname)s- %(message)s',
    handlers= [RotatingFileHandler(
       'appp.log',
       maxBytes=1*1024*1024,
       backupCount=5

    ),logging.StreamHandler()],
    force=True
)
logging.debug('This is a Debug message')
logging.info('This is a info message')
logging.warning('This is a warning message')
logging.error('This is a error message')
logging.critical('This is a critical message')
```

2. Implement TimedRotatingFileHandler to rotate logs daily at midnight. 

```python
from logging.handlers import TimedRotatingFileHandler

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        TimedRotatingFileHandler(
            'appp2.log',
            when='midnight',
            interval = 1,
            backupCount=7
        ),
        logging.StreamHandler()

    ],
    force=True
)

logging.info("Application started")
logging.debug("This is a debug message")
```

5. Write a multi-threaded program where each thread logs its name and
task progress safely. 

```python
import threading
import time
from config import logging

def task():
    for i in range(1,6):
        logging.info(f'Working...step {i+1}')
        # time.sleep(1)
# creating thread 
t1 = threading.Thread(target=task,name='task -1')
t2 = threading.Thread(target=task,name='thread -2')

# start thread
t1.start()
t2.start()

# wait thread for each other
t1.join()
t2.join()

logging.info('All task complete')
```
