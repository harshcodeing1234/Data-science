**Filtering :** Use a comparison inside square brackets:

```python
Employee = {
    "Name":['Dhananjay kumar pathak','Aarav pal','Harsh kumar','Prince Kumar','Om kuamr','Vaibhav kumar','Anmol Ray','Rohan Kumar','Tannu Kumari','Prashant Kumar','Rishi Ranjan'],
    "Age":[22,22,19,18,20,19,21,23,18,22,19],
    "ID":[4001,4002,4003,4004,4005,4006,4007,4008,4009,4010,4011],
    "Company":"Infosys",
    "Salery":[25000,35000,85000,75000,92000,56000,34000,67000,78000,45000,45500]

}

# Display data
df = pd.DataFrame(Employee)
print(df)

# Filter Age which are greater than 20
print("\n\nFilter Age Which are greater than 20")
filter_age = df[df['Age']>20]
print(filter_age)
```

With multiple conditions. 

- Use & for AND

- Use | for OR

- Each condition must be in parentheses.

```python
# Filtering with multiple conditions
print("FIltering Age and salery")

filterd = df[(df['Age']>20) & (df['Salery']>50000)]
print(filterd)

# OR conditions
print("\n\nUsing OR Conditions")

filterd = df[(df['Age']>20) | (df['Salery']>50000)]
print(filterd)

```
