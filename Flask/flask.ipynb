1. Create a Flask application that runs on port 5000 and displays the text “Hello, Flask
World!” when the root URL (/) is accessed.

```python
app = Flask(__name__)

@app.route('/')
def home():
    return "<h1>Hello</h1>"

if __name__ == '__main__':
    app.run()

```

2. Create a Flask app with three routes:
- / → displays “Home Page”
- /about → displays “About Us”
- /contact → displays “Contact Page”

```python
app = Flask(__name__)

@app.route('/')
def home():
    return f"THis is a Home Page"

@app.route('/about')
def about():
    return f'This is a about page'

@app.route('/contact')
def contact():
    return "This is a contact page"

app.run()
```

3. Create a dynamic route /user/<username> that displays the message “Welcome,
<username>!” where the username is taken from the URL

```python
app = Flask(__name__)

@app.route('/user/<username>')
def user(username):
    return f'<h1>Welcome {username}</h1>'

app.run()
```

4. Create a route /square/<int:number> that returns the square of the given number.

```python
app = Flask(__name__)

@app.route('/square/<int:number>')
def square(number):
    result = number*number
    return f'<h1>Square of {str(number)} = {str(result)}</h1>'

app.run()
```

6. Create anAPI endpoint /api/info that returns a JSON response containing a name,
version, and status.

```python
app = Flask(__name__)

@app.route('/api/info', methods=['GET'])
def api_info():
    return jsonify({
        'name':'Flask API',
        'version':'1.0000',
        'Status':'Running'
    })

app.run()
```

10. Create a route /home that redirects the user to /dashboard using url_for().


```python
app = Flask(__name__)

@app.route('/home')
def home():
    return redirect(url_for('dashboard'))

@app.route('/dashboard')
def dashboard():
    return "<h1>Welcome to dashboard</h1>"

app.run()
```
