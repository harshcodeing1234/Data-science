## Question 1: Average of List
WAP to input list and print the average

```python
num = input("Enter a list of numbers seprated by space:")
num_list = list(map(int,num.split()))
avrage = sum(num_list)/len(num_list)
print("average is",avrage)
print(f"Orignal list: {num_list}")
```

## Question 2: Square of Each Number
WAP to input a list of number and print the square of each number.

```python
num = input("Enter a list of numbers seprated by space:")
num_list = list(map(int,num.split()))
square = [x**2 for x in num_list]
print("square is", square)
```

## Question 3: Palindrome List Check
WAP to check list contains a palindrome or not

```python
lst1 = [1,2,1]
if(lst1[::-1] ==lst1):
	print("list are palindrome")
else:
	print("list are not palindrome")
```

## Question 4: Max and Min Numbers
WAP to input a list of number and print longest and smallest number

```python
num = input("Enter a list of numbers seprated by space:")
num_list = list(map(int,num.split()))
max_num = max(num_list)
min_num = min(num_list)
print("max number is",max_num)
print("min number is",min_num)
```

## Question 5: Student Marks
WAP to accept marks of 6 students and display them in a stored manner.

```python
marks = input("Enter marks seprated by space:")
marks_list = list(map(int,marks.split()))
print(f"marks of 6 student is {marks_list}")
```

## Question 6: Longest Name
WAP to input list of name and print longest name.

```python
name = input("Enter a list of names seprated by space:")
name_list = list(map(str,name.split()))
longest_name = max(name_list, key=len)
print("longest name is", longest_name)
```

## Question 8: Even and Odd Numbers
WAP to input list of number and print even and odd number.

```python
num = input("Enter a list of numbers seprated by space:")
num_list = list(map(int,num.split()))
for x in num_list:
	if x % 2 == 0:
		print(f"{x} is even")
	else:
		print(f"{x} is odd")
```
