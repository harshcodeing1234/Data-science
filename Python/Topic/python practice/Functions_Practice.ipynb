## Question 3: Even or Odd Check

```python
def even_odd(num):
    if num % 2 == 0:
        print("Even")
    else:
        print("odd")
even_odd(12)
even_odd(123456)
even_odd(123456789987654321)
```

## Question 4: Factorial
WAF to find factorial of n.

```python
def cal_fact(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    print(fact)
cal_fact(5)
```

## Question 5: Greeting Function
Write a program to greet a user with "Good day" using functions.

```python
import time
def greet():
    current_time = time.strftime('%H:%M %p')

    if current_time < "10:00 AM":
        print("Good day")
    else:
        print(current_time)

greet()
```

## Question 6: Fibonacci Sequence
WAF to find fibonacci sequence

```python
def fabonacchi(n):
    i = 0
    a =0
    b = 1  
    while i < n:
        print(a)
        a,b = b,a+b
        i = i+1
fabonacchi(10)
```

## Question 7: Greatest of Three
WAF to find greatest of three

```python
def check_greatest(a,b,c):
    if a>b and a>c:
        print("A is greatest")
    elif b>a and b>c:
        print("B is greatest")
    elif a ==b and a ==c:
        print("All are equal")
    else:
        print("C is greatest")
check_greatest(12,13,56)
check_greatest(1234523456345665,234567876543,3456789876543)
```

## Question 11: Sum of List Elements
WAF to takes list and returns sum of its element.

```python
num=input("Enter list number by space:")
num = list(map(int, num.split()))
def list_add(list):
    sum = 0
    for i in range(len(list)):
        sum = sum + list[i] 
    print(f"sum of list is:{sum}")
list_add(num)
```

## Question 14: Palindrome Check
WAF to check if a string is palindrome or not.

```python
def check_pal(str):
    if str == str[::-1]:
        print("String is palllindrome")
    else:
        print("String is not palindrome")

check_pal("Harsh")
```
