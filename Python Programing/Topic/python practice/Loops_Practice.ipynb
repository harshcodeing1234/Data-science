## Question 1: Hello World 10 Times
Print "Hello world" 10 times

````python
num = 1
while (num <=10):
    print("Hello World!")
    num = num+1
print("\n")
# 
for i in range(1,11):
    print("Hello World!")
````

## Question 2: Reverse Order Numbers
Print numbers from 10 to 1 in reverse order

````python
for i in range(10,0,-1):
    print(i)
````

## Question 3: Search Number in Tuple
Search a number x in tuple

````python
x = int(input("Enter number:"))
num = (1,4,9,16,25,36,49,64,81,100)

for i in num:
    if x == i:
        print(f"{x} is found")
        break
else:
    print(f"{x} not found")
````

## Question 4: Skip Number 5
Print numbers from 1 to 10 but skip 5

````python
for i in range(1,11):
    if i == 5:
        continue
    
    else:
        print(i)
````

## Question 5: Stop at 7
Print numbers from 1 to 50 but stop at 7

````python
i = 1
while (i <=50):
    if i == 7:
        break
    else:
        print(i)
    i = i+1
print("\n")
for i in range(1,51):
    if i == 7:
        break
    else:
        print(i)
````

## Question 6: Sum of Natural Numbers
Sum of first n natural numbers using while loop

````python
n = int(input("Enter num:"))

sum = 0
i = 1

while(i<n):
    sum = sum+i
    i = i+1
print(f"Sum of natural number is {sum}")
````

## Question 7: Odd and Even (While Loop)
Print odd and even numbers using while loop

````python
i = 1

while(i<20):
    if i%2 ==0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
    i = i+1
````

## Question 8: Odd and Even (For Loop)
Print odd and even numbers using for loop

````python
for i in range(1,21):
    if i%2!=0:
        print("odd = ",i)
    else:
        print("Even = ",i)
````

## Question 9: Multiplication Table (While Loop)
Multiplication table using while loop

````python
num = int(input("Enter num:"))

i = 1
while (i<=20):
    print(f"{num}*{i} = {num*i}")
    i = i+1
````

## Question 10: Multiplication Table (Reverse)
Multiplication table in reverse order

````python
num = int(input("Enter num:")) 
for i in range(10,0,-1):
    print(i,"*",num ,"=", i*num)
````

## Question 11: Multiplication Table (For Loop)
Multiplication table using for loop

````python
num = int(input("Enter number:"))
for i in range(1,11):
    print(num*i)
````

## Question 12: Greet Names Starting with S
Greet all person names starting with S

````python
l = ["Harry", "Soham", "Sachin", "Rahul"]
for name in l:
    if name.startswith('S'):
        print(f"Hello {name}")
````

## Question 13: Fibonacci Series
Print fibonacci series up to 10 terms

````python
i = 0
a , b = 0 , 1

while i < 10:
    print(a)
    a, b = b, a + b
    i = i + 1
````

## Question 14: Factorial
Calculate factorial using for loop

````python
num = int(input("Enter num:"))
fact = 1
for i in range(1,num+1):
    fact = fact*i
print(fact)
````
