## Question 3: Generate Multiplication Tables
Write a program to generate multiplication tables from 2 to 20 and write it to the different files. Place these files in a folder for a 13 â€“ year old.

## Question 4: Replace Word in File
Replace every occurrence of "Donkey" with "#####" in the file "word.txt"

```python
with open("word.txt", "r", encoding="utf-8") as f:
    content = f.read()

newcontent = content.replace("donkey", "######")

with open("word.txt", "w", encoding="utf-8") as f:
    f.write(newcontent)
```

## Question 7: Compare Files
Write a program to find out whether a file is identical & matches the content of another file.

```python
file1 = input("Enter file name:")
file2 = input("Enter file name:")
with open(file1, "r", encoding="utf-8") as f:
    content1 = f.read()
with open(file2, "r", encoding="utf-8") as f:
    content2 = f.read()

if content1 == content2:
    print("Yes both file is identical")
else:
    print("No files are not identical")
```
