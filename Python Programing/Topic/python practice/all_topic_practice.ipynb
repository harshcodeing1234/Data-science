WAF to check if a number is prime or not!

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input())
print(is_prime(n))
```

Given 3x3 matrix and calculte its transpose


```python
import numpy as np

def transpose(l1,l2,l3):
    if len(l1)!=3 or len(l2)!=3 or len(l3)!=3:
        raise IndexError("Only 3 element required")
    else:
        matrix = np.array([l1,l2,l3])
        transpose_matrix = matrix.T

        print(f"Orignal matrix:\n {matrix}")
        print(f"transpose matrix:\n {transpose_matrix}")

transpose([2,4,6],[3,4,5],[4,5,6])
```

Create a class `bankAccount` with **private** attribute _balance. implement method `deposit()` `withdrawal()` and a method to get the current balance ensuring the balance never goes negative .

```python
class bankAccount:
    def __init__(self,Account_holder,Account_no,balence=0):
        self.Accountholder = Account_holder
        self.Account_no = Account_no
        self.__balence = balence

    
    def deposite(self,Amount):
        self.__balence = self.__balence + Amount
        print(f"₹{Amount} deposite successfully")
        print(f"Your Current balence is {self.__balence}")

    
    def withrawl(self,Amount):
        if Amount > self.__balence:
            print("Insufficient balance! Withdrawal denied.")
        else:
            self.__balence = self.__balence - Amount
            print(f"₹{Amount} Widhrawll successfully")
            print(f"Your Current balence is {self.__balence}")

    def show_balence(self):
        print(f"Your total balence is {self.__balence}")

SBi = bankAccount('Harsh kumar',100491030050501)

SBi.deposite(2000)
SBi.withrawl(500)

SBI = bankAccount('shalini kumari',12344,6205808582)
SBI.deposite(1)
SBI.withrawl(1)
```

write a class student that initializes with name and grade(a list). Add a method to calculate the average grade

```python
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    
    def avrage(self):
        return sum(self.marks)/len(self.marks)
    
    def show(self):
        print(f"Average marks obtained by {self.name}: {self.avrage()}")

s1 = Student('Harsh',[23,45,67,89])
s1.show()
```

Create a vector class representing a 2D point(x,y). implementation the _str__ method to dispaly it as "(x,y)" and the  _add__ method to allow adding two vectors objects together using + operator

```python
class vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    def __add__(self,other):
        return vector(self.x+other.x,self.y+other.y)
    def __str__(self):
        return f"Vector: ({self.x},{self.y})"
    
v1 = vector(2,3)
v2 = vector(4,5)

add = v1+v2

print(add)
```

Create a class Calculator. Add a static method add(A,B ) that add Two numbers and a class method info() that return a string description of the class

```python
class calculator:

    @staticmethod
    def add(a,b):
        return a+b
    
    @classmethod
    def info(cls):
        return f"{cls.__name__} is a simple calculator class that performs basic operations."
    
result = calculator.add(12,20)
print(f"Addition: {result}")

print(calculator.info())
```
