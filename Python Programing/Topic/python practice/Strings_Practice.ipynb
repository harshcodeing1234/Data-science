## Question 1: Avoid Space in String Length
Calculate string length without counting spaces

```python
# Method 1: Using strip()
str = "   hello    "
print("Original length:", len(str))
print("After strip:", str.strip())
print("Length after strip:", len(str.strip()))

# Method 2: Using replace()
s = "    hello   "
res = len(s.replace(" ", ""))
print("Length without spaces:", res)
```

## Question 2: Print Even Length Word
Print even length word in a string

```python
str = "Hello World" 
# Print only even length words
if(len(str)) %2 == 0:
    print(str)
elif (len(str)) %2 == 1:
    print(str, end="!")
    print("\n")
```

## Question 4: Check Letter and Number
Check if a string has at least one letter and one number

```python
s = "abc123"

letter = any(char.isalpha() for char in s)
digit = any(char.isdigit() for char in s)

if letter and digit:
    print("String has at least one letter and one number")
else:
    print("String does not have both letter and number")
```

## Question 5: Remove Letters From String
Remove specific letters from a string

```python
str = "hello world" # Remove 'w'
print("Original string:", str)
print("After removing 'w':", str.replace("w" , ""))
```
