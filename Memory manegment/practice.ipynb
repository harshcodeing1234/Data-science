4. Create a program that generates circular references between two objects and show how
Python’s garbage collector handles them using the gc module. 

```python
import gc

class A:
    pass

class B:
    pass

a = A()
b =B()

a.b = b
b.a = a

del a
del b

print(gc.collect())
```

7. Write a program that compares memory addresses when using assignment (=),
copy.copy(), and copy.deepcopy() on a nested list

```python
import copy

original = [[1, 2], [3, 4]]

assigned = original
shallow = copy.copy(original)
deep = copy.deepcopy(original)

print("Original:", id(original))
print("Assigned:", id(assigned))
print("Shallow:", id(shallow))
print("Deep:", id(deep))

print("Inner list IDs:")
print(id(original[0]), id(shallow[0]), id(deep[0]))
```
