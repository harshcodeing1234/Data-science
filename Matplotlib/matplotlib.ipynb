**Matplotlib**

**What is Matplotlib** :- Matplotlib is a low level graph plotting library in python that serves as a visualization utility.
- Matplotlib was created by John D. Hunter.
- Matplotlib is open source and we can use it freely.

- Import Matplotlib

```python
import matplotlib
print(matplotlib.__version__)
```

**Pyplot** :- Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

Example:

```python
x = [1,2,3,4]
y = [10,2,12,40]
z = [23,45,21,44]
plt.plot(x,z,)  
plt.show() # method to show 
```

**All functions in plot** :

| Function                     | Purpose                          | Example                              |
| ---------------------------- | -------------------------------- | ------------------------------------ |
| `plt.plot()`                 | Plot data (line graph)           | `plt.plot(x, y)`                     |
| `plt.xlabel()`               | Label for X-axis                 | `plt.xlabel("Time (s)")`             |
| `plt.ylabel()`               | Label for Y-axis                 | `plt.ylabel("Speed (m/s)")`          |
| `plt.title()`                | Add a title to graph             | `plt.title("Speed vs Time")`         |
| `plt.grid()`                 | Show grid lines                  | `plt.grid(True)`                     |
| `plt.legend()`               | Display legend for labeled lines | `plt.legend()`                       |
| `plt.xlim(), plt.ylim()`     | Set limits for axes              | `plt.xlim(0,10)`                     |
| `plt.xticks(), plt.yticks()` | Set tick positions and labels    | `plt.xticks([1,2,3], ['A','B','C'])` |
| `plt.show()`                 | Display the plot window          | `plt.show()`                         |


- Mark `lable, title` in charts

```python
x = ["Mon", "Tue", "Wed", "Thursday", "Fri", "Sat"]
y = [10, 12, 8, 56, 21, 5]

plt.plot(x,y)
plt.title("Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales per day")

plt.show()
```

- Add grid in background

```python
x = ["Mon", "Tue", "Wed", "Thursday", "Fri", "Sat"]
y = [1, 4, 3, 6, 2, 5]

plt.plot(x,y, color='blue', linestyle='--', marker='o', linewidth=2, label="2025 sales data") # All parameter
plt.title("Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales per day")
plt.grid(color='gray', linestyle=':', linewidth='1') # All parameter
plt.legend(loc="upper left") # Show lable that are passed in plot.
plt.show()
```

- Set limit in x and y

```python
x = [10,20,30,40,50,60]
y = [1, 4, 3, 6, 2, 5]

plt.plot(x,y)
plt.title("Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales per day")
plt.grid()
plt.xlim(10,40) #pass range
plt.ylim(2,5)  # pass range
plt.show()
```

- `plt.xticks()` and `plt.yticks()` : `xticks` and `yticks` are used to control the tick locations and labels on the `X` and `Y` axes of a plot.

```python
x = [5,10,15,20]
y = [23,21,22,34]

plt.plot(x,y)
plt.xticks([5,10,15,20], ['A', 'B', 'C', 'D'])
plt.yticks([23,21,22,34], [23,21,22,24])

plt.show()
```

**Bar charts** : A bar chart (or bar graph) is a graphical representation of data using rectangular bars, where the length or height of each bar represents the value or frequency of a category.

```python
product = ["Oil","Shoap","Pen","Shampo","Powder"]
sales = [60,20,45,32,56]

plt.bar(product, sales, color='orange', label='Today Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title("Today sales statics")
plt.legend()
plt.show()
```

**Pie Charts** : A pie chart is a circular graph that shows data as slices of a circle, where each slice represents a proportion or percentage of the whole.

```python
country = ['India','Japan','Srilanka','America','Dubai']
Sales = [78,67,45,91,88]
colors=['orange','blue','red','yellow','green']
explode = (0.2,0,0,0,0)
plt.pie(sales, labels=country,explode=explode, autopct='%1.1f%%',colors=colors,shadow=True)
plt.title('Revenue contribute by countries')
plt.show()
```

**histograms** : A histogram shows how data is distributed  it groups values into bins and displays their frequency.

```python
marks = [22, 55, 62, 45, 21, 34, 42, 42, 55, 62, 68, 70, 75, 80, 85, 90, 95]

plt.hist(marks,bins=4,color='yellow',edgecolor='black')
plt.xlabel('Marks Range')
plt.ylabel("No. of students")
plt.title('Student marks')
plt.show()
```

**Scatter plot** : A scatter plot is a type of graph in which each point represents the value of two variables  one plotted along the x-axis and the other along the y-axis

```python
study_hours = [2,4,6,8,10,12]
marks = [55,70,80,85,90,94]

plt.scatter(study_hours, marks, color='blue', marker='d', label='Student Data')
plt.xlabel('Study hours')
plt.ylabel('Total marks')
plt.title('Relationship between study hours and marks')
plt.grid()
plt.legend()
plt.show()
```

- Example with Multiple Datasets:

```python
plt.scatter([2,3,5,7,8],[55,60,67,89,92],color='blue',marker='o',label='Sec "A"')
plt.scatter([1,4,6,8,9],[45,67,78,88,90],color='orange',marker='s',label='Sec "B"')
plt.xlabel('Study hours')
plt.ylabel('Total marks')
plt.title('Comparision between Sec A and Sec B')
plt.grid()
plt.legend()
plt.show()
```

**Subplot** : A subplot allows you to display multiple plots in one figure window  useful for comparing different visualizations side by side or in a grid layout.
plt.subplot(nrows, ncols, index)


```python
import matplotlib.pyplot as plt

# Data
study_hours = [2, 4, 6, 8, 10, 12]
marks = [55, 70, 80, 85, 90, 94]

# 1st subplot (Scatter Plot)
plt.subplot(1, 3, 1)   # 1 row, 3 columns, 1st plot
plt.scatter(study_hours, marks, color='blue', marker='o')
plt.title('Scatter Plot')
plt.xlabel('Study Hours')
plt.ylabel('Marks')

# 2nd subplot (Line Plot)
plt.subplot(1, 3, 2)   # 1 row, 3 columns, 2nd plot
plt.plot(study_hours, marks, color='green', marker='s')
plt.title('Line Plot')
plt.xlabel('Study Hours')
plt.ylabel('Marks')

# 3rd subplot (Bar Plot)
plt.subplot(1, 3, 3)   # 1 row, 3 columns, 3rd plot
plt.bar(study_hours, marks, color='orange')
plt.title('Bar Plot')
plt.xlabel('Study Hours')
plt.ylabel('Marks')

plt.tight_layout()  # Adjust spacing
plt.show()
```

**using fig, ax = `plt.subplots()`** :
 
- fig (Figure): The overall window or canvas where all your plots live.

- ax (Axes): The actual plotting area  where you draw one chart.

```python
# Data
x = [12, 34, 56, 78]
y = [100, 345, 678, 900]

# Create figure and subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Bar Chart
ax[0].bar(x, y, color='orange')
ax[0].set_title('Reports')
ax[0].set_xlabel('X Values')
ax[0].set_ylabel('Y Values')

# Pie Chart
ax[1].pie(y, labels=x, autopct='%1.1f%%', colors=['yellow', 'red', 'blue', 'orange'])
ax[1].set_title('Pie Reports')

# Add overall title
fig.suptitle("All Reports", fontsize=14)

# Adjust spacing
plt.tight_layout()
plt.show()
```

`savefig()` : `savefig()` is used to save your current figure (the chart or plots youve created) as an image file  like PNG, JPG, PDF, SVG, etc.

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 45]

plt.plot(x, y, color='green', marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Save before showing
plt.savefig('line_plot_highres.png', dpi=300, bbox_inches='tight')
plt.show()
```
