```python
data = {
    'Month':['January','February','March','April','May','Jume','July','August','September','October','November','December'],
    'Sales':[80000,78000,67000,56000,67000,89000,120000,96000,72000,10000,34000,12000]
}

df = pd.DataFrame(data)
df.to_csv('sales_basic.csv',index=False)
```

Create a line plot 


```python
plt.figure(figsize=(25,12))
plt.plot(df['Month'],df['Sales'],marker='o',color='purple')
plt.grid()
plt.show()
```

Add X and Y axis labels 


```python
plt.figure(figsize=(22,8))
plt.plot(df['Month'],df['Sales'],marker='o',color='purple')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

Add a title 


```python
plt.figure(figsize=(22,8))
plt.plot(df['Month'],df['Sales'],marker='o',color='purple')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales over Month')
plt.grid()
plt.show()
```
