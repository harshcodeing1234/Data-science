# Matplotlib & Seaborn Practice

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Level 01: Basic**

- Load the dataset and display the first 10 rows.
- Check for null values in each column.
- Display basic statistics (mean, median, min, max) for numerical columns.

```python
df = pd.read_csv('matplotlib_seaborn_dataset.csv')
print(df.head(10))
df.describe()
```

```python
df.isnull().sum()
```

```python
Date = df['Date'].head(10)
Sales = df['Sales'].head(10)
profit = df['Profit'].head(10)
city = df['City'].head(10)
```

Plot a line chart of Temperature over Date.


```python
plt.figure(figsize=(25,12))
plt.plot(Date,Sales,color='purple',marker='o',label='Sales over Date')
plt.title('Sales over Date')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend(loc='upper left')
plt.show()
```

Create a bar chart showing counts of the City.


```python
city = df['City'].value_counts()
plt.bar(city.index,city.values,color='yellow',label='Counts of the city')
plt.legend()
plt.show()
```

Draw a histogram of Sales.


```python
plt.Figure(figsize=(12,6))
plt.hist(df['Sales'],bins=5,color='purple',edgecolor='black')
plt.title('Histogram of Sales')
plt.show()
```

Plot a pie chart of Weather distribution and select hiest weather

```python
weather = df['Weather'].value_counts()
color=['red','purple','yellow','green']
explocode = [0.01,0.2,0.01,0.01]
plt.pie(weather,labels=weather.index,colors=color,autopct='%1.1f%%',explode=explocode)
plt.title('Weather distribution')
plt.show()
```

Create a scatter plot between Temperature and Humidity.


```python
plt.figure(figsize=(25,12))
plt.scatter(df['Temperature'],df['Humidity'],marker='*',edgecolors='black')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.grid()
plt.title('Scatter plot between Temperatature and Humidity')
plt.show()
```

Plot Profit over Date using a line chart.


```python
plt.figure(figsize=(25,12))
plt.plot(Date,profit,marker='o',color='red',linestyle='--')
plt.xlabel('Profit')
plt.ylabel('Date')
plt.title('Scatter plot between Profit and Date')
plt.show()
```

Display top 5 City by total Sales using a bar chart.


```python
df_groupd = df.groupby('City')['Sales'].sum().nlargest(5)
plt.bar(df_groupd.index,df_groupd.values,color='purple',label='Top 5 city by total saled')
plt.legend()
plt.show()
```

Count the number of occurrences of each Weather type.


```python
Weather = df['Weather'].value_counts()
plt.bar(weather.index,weather.values,color='green',label='Occurences of weather')
plt.legend()
plt.show()
```

Show a density plot of Profit.

```python
sns.kdeplot(profit,fill=True)
xlabel=('Profit')
ylabel=('Density')
plt.show()
```

Plot Temperature vs Sales using a scatter plot.


```python
plt.scatter(df['Temperature'],df['Sales'],marker='x')
plt.show()
```

Draw a histogram for Humidity with 20 bins.

```python
plt.hist(df['Humidity'],bins=20,edgecolor='black')
plt.show()
```

**Level 02: Medium level**

Handle missing values by imputing with mean for Temperature.


```python
df.fillna(df['Temperature'].mean(),inplace=True)

```

Create a box plot of Profit grouped by City.


```python
City=city.unique()
sns.boxplot(x='City',y='Profit',data=df)
plt.title('Box plot of City over Profit')
plt.show()

```

Draw a violin plot of Sales by Weather.


```python
sns.violinplot(x=df['Weather'],y=df['Sales'])
plt.title('Voilen plot of weather over sales')
plt.show()
```

Create a heatmap of correlation among numerical columns

```python
numeric_feature =['Temperature','Humidity','Sales','Profit']
corr = df[numeric_feature].corr()

# visualize
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Heatmap of correlation between numeric features')
plt.show()
```

Create subplots for Temperature, Humidity, and Profit over Date.


```python

plt.figure(figsize=(12, 6))

# 1st subplot  Temperature
plt.subplot(3, 1, 1)
plt.plot(df['Date'], df['Temperature'], color='green')
plt.title('Temperature Over Date')
plt.ylabel('Temperature')

# 2nd subplot  Humidity
plt.subplot(3, 1, 2)
plt.plot(df['Date'], df['Humidity'], color='blue')
plt.title('Humidity Over Date')
plt.ylabel('Humidity')

# 3rd subplot  Profit
plt.subplot(3, 1, 3)
plt.plot(df['Date'], df['Profit'], color='red')
plt.title('Profit Over Date')
plt.xlabel('Date')
plt.ylabel('Profit')

plt.tight_layout()
plt.show()
```

Group data by City and plot mean Profit in a bar chart.


```python
grouped = df.groupby('City')['Profit'].mean()
plt.bar(grouped.index,grouped.values,color='orange',label='Average profit by city')
plt.legend()
plt.show()
```

Draw a box plot of Temperature with horizontal orientation.


```python
sns.boxplot(x='Profit',y='Weather',data=df)
plt.title('Box plot of weather over profit')
plt.show()
```

Use hue to show Weather categories in scatter plots.


```python
sns.scatterplot(x='Temperature',y='Sales',hue='Weather',data=df)
plt.title('Scatter plot of Temperature vs Sales colored by Weather')
plt.show()
```

Plot `Sales` distribution by `City` using KDE plots.


```python
sns.kdeplot(df['Sales'],fill=True)
plt.title('KDE plot of sales')
plt.show()
```

Plot Profit vs Sales with regression line using Seaborn's regplot.


```python
sns.regplot(x='Sales',y='Profit',data=df,ci=None)
plt.title('Regression plot between Profit and Sales')
plt.grid()
plt.show()
```

Draw a line plot of Humidity and overlay a rolling mean.


```python
rolling = df['Humidity'].rolling(window=5,min_periods=1).mean()
plt.plot(rolling,label='Rolling mean(5)')
plt.legend()
plt.title('Humidity With rolling mean')
plt.show()
```

```python
# 34. Draw a line plot of `Humidity` and overlay a rolling mean.
rolling = df['Humidity'].rolling(window=5,min_periods=1).mean()
plt.plot(rolling,label = 'Rolling Mean (5)')
plt.xlabel('Index')
plt.ylabel('Humidity')
plt.title("Humidity With rolling Mean")
plt.legend()
plt.show()
```
