### Multithreading practice 

```python
import threading
```

1. Write a Python program to create two threads, where one thread prints numbers
from 1 to 10 and the other prints numbers from 11 to 20. 

```python
def print_num():
    for i in range(1,11):
        print(f'Number: {i}')

def print_nums():
    for i in range(11,21):
        print(f'number: {i}')

# Create thread
t1 = threading.Thread(target=print_num)
t2 = threading.Thread(target=print_nums)

# Start the thread
t1.start()
t2.start()

# Wait for the thread to complete
t1.join()
t2.join()
```

2. Create a multithreaded program in which one thread prints even numbers and
another thread prints odd numbers up to 50. 

```python
def print_even(num):
    for i in range(2,num+1,2):
        print(f'Even_num {i}')

def odd_num(num):
    for i in range(1,num+1,2):
        print(f'Odd_num {i}')
num=20
# Create thread
t1 = threading.Thread(target=print_even,args=(num,))
t2 = threading.Thread(target=odd_num,args=(num,))

# Start the thread
t1.start()
t2.start()

# Wait for the thread to complete
t1.join()
t2.join()
```

3. Write a program using multithreading to calculate the square and cube of a given
number simultaneously

```python
def square_num():
    for i in range(1,11):
        result = f'Square of {i} is {i*i}'
        print(result)

def cube_num():
    for i in range(1,11):
        result = f'Cube of {i} is {i*i*i}'
        print(result)

# Create thread
t1 = threading.Thread(target=square_num)
t2 = threading.Thread(target=cube_num)

# Start the thread
t1.start()
t2.start()

# Wait for the thread to complete
t1.join()
t2.join()
```
