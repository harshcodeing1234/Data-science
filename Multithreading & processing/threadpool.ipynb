### ThreadPoll Practice

```python
import concurrent.futures as cf
import requests
```

1. Write a Python program using ThreadPoolExecutor to download multiple web
pages concurrently. 

```python
urls = [
    "https://www.example.com",
    "https://www.w3.org",
    "https://www.python.org"
]

def download_page(url):
    response = requests.get(url)
    print(f'{url} downloaded! , length = {len(response.text)} characters')

# using threadpollexecuter 
with cf.ThreadPoolExecutor(max_workers=3) as executor:
    executor.map(download_page,urls)
```

2. Create a program using a thread pool that executes multiple I/O-bound tasks and
prints the thread name for each task. 

```python
import concurrent.futures
import threading
import time

# I/O-bound task (simulated using sleep)
def io_task(task_id):
    print(f"Task {task_id} started by {threading.current_thread().name}")
    time.sleep(2)  # Simulate I/O delay
    print(f"Task {task_id} finished by {threading.current_thread().name}")

tasks = [1, 2, 3, 4, 5]

# Using ThreadPoolExecutor
with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
    executor.map(io_task, tasks)
```
